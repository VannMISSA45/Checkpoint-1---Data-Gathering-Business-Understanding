# Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Membaca Data

```python
df = pd.read_csv("/content/train.csv")
df
```

```python

df_train.columns = ['File_Name', 'Action', 'Value']
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df_train['Action'].value_counts().index, df_train['Action'].value_counts().values, color='blue')
plt.title('Jumlah Tindakan Berdasarkan Kategori')
plt.xlabel('Kategori Tindakan')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Action', data=df_train, hue="Value", palette='pastel')
plt.title('Jumlah Tindakan Berdasarkan Kategori dan Nilai')
plt.xlabel('Kategori Tindakan')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(df_train['Action'].value_counts(), labels=df_train['Action'].value_counts().index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue', 'lightgreen', 'gold'])
plt.title('Distribusi Kategori Tindakan')
plt.show()
```
